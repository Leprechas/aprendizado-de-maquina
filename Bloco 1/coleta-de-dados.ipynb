Como pode ser observado, a base de dados utilizada provê informações de expressão gênica de um determinado gene para diferentes amostras. Em um segundo banco de dados, cada amostra indica o sexo do indivíduo doador da amostra, assim como seu intervalo de idade e um terceiro dado que não conseguimos identificar.

```python
sample_data = pd.read_csv('sample.csv')
```

```python
sample_data.head()
```

```python
sample_data = sample_data.rename(columns={"SUBJID": "Description"})
```

```python
modified_data = raw_data.reset_index(drop=True)
modified_data = modified_data.set_index('Description')
modified_data = modified_data.T
modified_data.head()
```

```python
modified_data = modified_data.reset_index()
modified_data = modified_data.rename(columns={modified_data.columns[0]:'id','index':'Description'})
modified_data.index.names = ['']
modified_data.head()
```

```python
modified_data = modified_data.rename(columns={'Description':'Sample'})
```

```python
sample_data = sample_data.rename(columns={'Description':'Sample'})
```

Uma vez que temos os *datasets* com as informações desejadas, basta agora encontrar a relação de cada um deles e realizar uma junção. Entretanto, é importante notar que enquanto na primeira base a amostra é dada com uma identificação do tecido (peles não expostas ao sol, neste caso), na segunda ela é dada apenas em função do doador, explicitado pelos 9 primeiros dígitos. Assim, nas células seguintes, o código completo da amostra é substituído apenas para o código do doador, possibilitando finalmente o *inner join* e exportação da base para um arquivo em *csv*.

```python
for i in range(len(modified_data['Sample'])):
    modified_data = modified_data.replace(modified_data['Sample'][i],modified_data['Sample'][i][:9])
```

```python
joined_data = pd.merge(modified_data,sample_data,on="Sample",how='inner')
```

```python
joined_data.to_csv('raw_data.csv')
```
